```python
#Exercicio 2
import numpy as np # Importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca
# sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
#Dados de treinamento 
calor = np.array([100, 200, 300, 400, 500, 600, 700]).reshape(-1,1)
#Quantidade de calor
temperatura = np.array([30,35,40,45, 60, 100, 150]) #Nível da temperatura
#Criar e treinar o modelo de regressão linear
modelo = LinearRegression() #Cria uma variavel chamada modelo
modelo.fit(calor, temperatura) #Cria o modelo com base nos dados de entrada
```

```python
#Dados de teste
#vazao_teste = np.array([15,25, 33]).reshape(-1,1) #vazao de agua em m3/min
calor_teste = np.array([500,900,150,370,90,55,70]).reshape(-1,1) #vazão de agua em m3/min
#Fazer previsões com base na vazão de água
nivel_previsto = modelo.predict(calor_teste)
```

```python
#Imprimir as previsões
for i in range(len(calor_teste)): #Laço de repetição para exibir os dados de saida
    print(f"Calor: {calor_teste[i][0]}  - Nível previsto : {nivel_previsto[i]} temperatura")
#grafico de pontos - scatter
plt.scatter(calor, temperatura, color='blue', label='Dados de Teste')
#Grafico de linhas
plt.plot(calor_teste, nivel_previsto, color='red', label='Dados de treinamento')
plt.xlabel('Calor')
plt.ylabel('Nível da temperatura')
plt.title('Relação entre calor e temperatura')
plt.legend()
plt.grid(True)
plt.show()
```
