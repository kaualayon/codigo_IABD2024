```python
#Exercicio 4
import numpy as np # Importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca
# sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
#Dados de treinamento 
renda = np.array([1200, 5000, 8000,10000, 30000]).reshape(-1,1)
#Quantidade de gastos
gastos = np.array([750, 1500, 4000, 6000, 18000]) #Nível da valor
#Criar e treinar o modelo de regressão linear
modelo = LinearRegression() #Cria uma variavel chamada modelo
modelo.fit(renda, gastos) #Cria o modelo com base nos dados de entrada
```

```python
#Dados de teste
renda_teste = np.array([25000, 6000, 10000,8000, 9000]).reshape(-1,1) #area
#Fazer previsões com base na area
gasto_previsto = modelo.predict(renda_teste)
```

```python
#Imprimir as previsões
for i in range(len(renda_teste)): #Laço de repetição para exibir os dados de saida
    print(f"Renda: {renda_teste[i][0]}  - Gasto previsto : {gasto_previsto[i]}")
#grafico de pontos - scatter
plt.scatter(renda, gastos, color='blue', label='Dados de Teste')
#Grafico de linhas
plt.plot(renda_teste, gasto_previsto, color='red', label='Dados de treinamento')
plt.xlabel('Renda')
plt.ylabel('Gasto previsto')
plt.title('Relação entre renda e gastos')
plt.legend()
plt.grid(True)
plt.show()
```
