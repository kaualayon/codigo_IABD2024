```python
#Exercicio 3
import numpy as np # Importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca
# sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
#Dados de treinamento 
area = np.array([50, 100, 120,300]).reshape(-1,1)
#Quantidade de calor
valor = np.array([180000, 300000, 375000, 600000]) #Nível da temperatura
#Criar e treinar o modelo de regressão linear
modelo = LinearRegression() #Cria uma variavel chamada modelo
modelo.fit(area, valor) #Cria o modelo com base nos dados de entrada
```

```python
#Dados de teste
area_teste = np.array([500,900,150,370,90,55,70]).reshape(-1,1) #area
#Fazer previsões com base na area
valor_previsto = modelo.predict(area_teste)
```

```python
#Imprimir as previsões
for i in range(len(area_teste)): #Laço de repetição para exibir os dados de saida
    print(f"Área: {area_teste[i][0]}  - Valor previsto : {valor_previsto[i]}")
#grafico de pontos - scatter
plt.scatter(area, valor, color='blue', label='Dados de Teste')
#Grafico de linhas
plt.plot(area_teste, valor_previsto, color='red', label='Dados de treinamento')
plt.xlabel('Area')
plt.ylabel('Valor previsto')
plt.title('Relação entre area e valor')
plt.legend()
plt.grid(True)
plt.show()
```
